```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.bikewale.com/ktm-bikes/"
```

```python
ktm=requests.get(url)
```

```python
ktm
```

```python
ktm=BeautifulSoup(ktm.text)
```

```python
ktm
```

```python
model=ktm.select('.front:nth-child(2) .modelurl , .front:nth-child(6) .modelurl , .front:nth-child(5) .modelurl , .front:nth-child(7) .modelurl , .front:nth-child(8) .modelurl')
```

```python
model
```

```python
price=ktm.select('li:nth-child(1) .font11+ .grid-6 .text-bold , #listitems .font18')
```

```python
price
```

```python
import pandas as ps
```

```python
ktmlist=ps.DataFrame(model,columns=["modelname"])
ktmlist["modelprice"]=price
```

```python
ktmlist
```
